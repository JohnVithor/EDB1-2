```python
# Importando os componentes necessários
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Carregando os dados a serem analisados

        # Organização Não Decrescente
tempo_V_ND = pd.read_csv('../tempos/tempo_V_ND.csv')
                        
        # Organização Não Crescente
tempo_V_NC = pd.read_csv('../tempos/tempo_V_NC.csv')
                        
        # Organização Aletória
tempo_V_A = pd.read_csv('../tempos/tempo_V_A.csv')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(13,13))
fig.tight_layout()
funcoes = ['Insertion','Selection','Bubble','Shell','Merge','Quick','Radix']

for i in range(0,7):
    ax=tempo_V_ND.plot(kind='line',x='tam_vet',y=funcoes[i],
                   title=funcoes[i],
                   color='blue', legend=False,
                   ax=axes[i//3, i%3])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_xlabel("Quantidade de Elementos")
    ax.set_ylabel("Tempo em Milisegundos")
plt.tight_layout()
fig.savefig('./Não_Decrescente.png')
```

```python
fig2, axes2 = plt.subplots(nrows=3, ncols=1, figsize=(13,13))

funcoes = ['Insertion','Selection','Bubble','Shell','Merge','Quick','Radix']
funcoes_L = ['Selection','Quick']
funcoes_R = ['Insertion','Bubble','Shell','Merge','Radix']

ax = tempo_V_ND.plot(kind='line',x='tam_vet',y=funcoes_L,title="Elementos em Ordem Não Decrescente",
                color=['green','red'], legend=True,ax=axes2[0])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

ax = tempo_V_ND.plot(kind='line',x='tam_vet',y=funcoes_R,title="Elementos em Ordem Não Decrescente",
                color=['blue','pink','yellow','orange','black'], legend=True,ax=axes2[1])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

ax = tempo_V_ND.plot(kind='line',x='tam_vet',y=[funcoes[0],funcoes[2]],title="Comparação entre Bubble sort e Insertion sort",
                color=['blue','pink'], legend=True,ax=axes2[2])
ax.set_xlabel("Quantidade de Elementos")
ax.set_ylabel("Tempo em Milisegundos")

plt.tight_layout()
fig2.savefig('./Não_Decrescente_Comparações.png')
```
